```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load Fashion MNIST dataset
fashion_mnist = keras.datasets.fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
# Normalize the data
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
# Check shapes
print("Training data shape:", x_train.shape)
print("Test data shape:", x_test.shape)
```

```python
# actual names of the 10 categories the model will predict.
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
# Reshape the data to add a channel dimension (since it's grayscale)
# CNNs expect a 4D input: (batch_size, height, width, channels)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
```

```python
# Build CNN model
model = keras.Sequential([
    layers.Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D(pool_size=(2, 2)),

    layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),   #(to learn image features).
    layers.MaxPooling2D(pool_size=(2, 2)),

    layers.Flatten(),           #converts image to a 1D array.
    layers.Dense(64, activation='relu'),    #make predictions.
    layers.Dense(10, activation='softmax')  # Output layer for 10 classes
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Display the model summary
model.summary()
```

```python
# Train the model
history = model.fit(x_train, y_train, epochs=10, validation_split=0.2)
```

```python
# Evaluate the model
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test accuracy: {test_acc:.4f}")
```

```python
# Plot training history
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
```

```python
# Making Predictions
y_pred = model.predict(x_test)
y_pred_classes = y_pred.argmax(axis=1)  # argmax picks the class with the highest probability.

# Print results
for i in range(5):
    print(f"{i+1} - Predicted: {y_pred_classes[i]}, Actual: {y_test[i]}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Predict on the full test set
y_pred = model.predict(x_test)

# Collect one image per class
images_per_class = []
labels_seen = set()

for i in range(len(x_test)):
    label = y_test[i]
    if label not in labels_seen:
        images_per_class.append((x_test[i], y_pred[i], label))
        labels_seen.add(label)
    if len(images_per_class) == 10:
        break

# Plot bigger images
plt.figure(figsize=(20, 6))  # <-- Increase width and height

for i, (image, pred, actual) in enumerate(images_per_class):
    plt.subplot(2, 5, i + 1)  # use 2 rows if needed, or stick with (1, 10)
    plt.imshow(image.reshape(28, 28), cmap='gray')
    plt.title(f"Pred: {class_names[np.argmax(pred)]}\nAct: {class_names[actual]}", fontsize=16)  # slightly bigger font
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

# Get predicted class labels
y_pred_classes = y_pred.argmax(axis=1)

# Classification report
print("Classification Report:\n")
print(classification_report(y_test, y_pred_classes, target_names=class_names))


```
